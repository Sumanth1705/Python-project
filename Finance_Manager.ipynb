```python
import pandas as pd
import matplotlib.pyplot as plt

# Define a class to manage finances
class FinanceManager:
    def __init__(self):
        self.transactions = pd.DataFrame(columns=["Date", "Amount", "Category", "Description"])

    def add_transaction(self, date, amount, category, description):
        self.transactions.loc[len(self.transactions)] = [date, amount, category, description]

    def get_balance(self):
        return self.transactions["Amount"].sum()

    def get_spending_by_category(self):
        return self.transactions.groupby("Category")["Amount"].sum()

    def plot_spending(self):
        spending_by_category = self.get_spending_by_category()
        plt.bar(spending_by_category.index, spending_by_category.values)
        plt.xlabel("Category")
        plt.ylabel("Amount")
        plt.title("Spending by Category")
        plt.show()

# Create a FinanceManager object
finance_manager = FinanceManager()

# Add some transactions
finance_manager.add_transaction("2023-01-01", 100, "Groceries", "Bought groceries for the week")
finance_manager.add_transaction("2023-01-02", 50, "Entertainment", "Went to the movies")
finance_manager.add_transaction("2023-01-03", 25, "Transportation", "Took the bus to work")

# Get the balance
balance = finance_manager.get_balance()
print(f"Current balance: {balance}")

# Get the spending by category
spending_by_category = finance_manager.get_spending_by_category()
print(f"Spending by category: {spending_by_category}")

# Plot the spending
finance_manager.plot_spending()
```
