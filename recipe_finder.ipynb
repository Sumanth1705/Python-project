```python
import csv

# Function to read recipes from CSV file
def read_recipes_from_csv(file_path):
    recipes = {}
    with open(file_path, 'r', newline='') as file:
        reader = csv.reader(file)
        for row in reader:
            ingredient, *recipe_list = row
            recipes[ingredient.lower()] = [recipe.strip() for recipe in recipe_list]
    return recipes

# Function to find recipes based on user input
def find_recipes(ingredient, recipes):
    ingredient = ingredient.lower()
    if ingredient in recipes:
        return recipes[ingredient]
    else:
        return []

# Main function
def main():
    print("Welcome to Recipe Finder!")
    # Specify the file path to your uploaded CSV file
    file_path = 'recipes.csv'

    recipes = read_recipes_from_csv(file_path)
    while True:
        ingredient = input("Enter an ingredient to find recipes (or 'quit' to exit): ")
        if ingredient.lower() == 'quit':
            print("Thank you for using Recipe Finder. Goodbye!")
            break
        recipes_found = find_recipes(ingredient, recipes)
        if recipes_found:
            print("Recipes found for", ingredient.capitalize() + ":")
            for recipe in recipes_found:
                print("-", recipe)
        else:
            print("No recipes found for", ingredient.capitalize())

# Execute the program
if __name__ == "__main__":
    main()
```
